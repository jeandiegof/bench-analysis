```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

## Read data

```python
df = pd.read_csv('cpu-freq.csv')
cores = 8

fig, ax = plt.subplots(cores, 1, figsize=(15, 15))

for i in range(8):
    sns.lineplot(x='sample', y='cpu{}'.format(i), data=df, ax=ax[i]);
```
